Realizamos una pequeña descripción de los datos númericos su media, desviación estandar, el valor minímo y máximo, los cuartiles, etc.

```python
moda_overall = df["overall"].mode
print(moda_overall)
varianza_edad = df["age"].var(ddof=0)
print("\nLa varianza en la edad es: ",varianza_edad)
desviacion_edad = df["age"].std(ddof=0)
print("\nLa desviación estandar en la edad es: ",desviacion_edad)

varianza_overall = df["overall"].var(ddof=0)
print("\nLa varianza del nievel de portero es: ",varianza_overall)
desviacion_overall = df["age"].std(ddof=0)
print("\nLa desviación estandar en el nivel es: ",desviacion_overall)

varianza_valor = df["value_eur"].var(ddof=0)
print("\nLa varianza del valor en euros de portero es: ",varianza_valor)
desviacion_valor = df["value_eur"].std(ddof=0)
print("\nLa desviación estandar del valor en euros del portero es: ",desviacion_valor)
```

calculamos la moda, la variazna y desviación estandar de algunos datos relevantes.

```python
sns.pairplot(df[[ 'value_eur','age','overall']],palette="pastel")
```

aquí comparamos tres variables diferentes con distintas gráficas nos da un resumen de concentraciuón de datos y de como se mueven cuando alguno de los tres datos comparados aumenta o disminuye

```python
sns.lmplot(x='overall', y='potential', hue='preferred_foot',
           data=df.loc[df['preferred_foot'].isin(['Left', 'Right'])],
           fit_reg=False,palette="husl")
```

en esta gráfica podemos ver que no influye tanto el pie preferído del portero con su nivel ya que podemos ver puntos de ambos colores por todos lados

Histogramas de la frecuencia del tipo de cuerpo en comparación con su nivel.

```python
m = plt.hist(df[df["body_type"] == "Normal"].overall,bins=30,fc = (1,0,0,0.5),label = "Normal")
b = plt.hist(df[df["body_type"] == "Lean"].overall,bins=30,fc = (0,1,0,0.5),label = "Delgado")
c = plt.hist(df[df["body_type"] == "Stocky"].overall,bins=30,fc = (0,0,1,0.5),label = "Stocky")


plt.legend()
plt.xlabel("Nivel")
plt.ylabel("Frecuencia")
plt.title("Histograma")
plt.show()
```

aquí vemos juntos donde se concetran la cantidad de porteros y la proporción que tienen en la frecuencia del tipo de cuerpo que tienen estos en cada nivel
